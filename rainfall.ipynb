```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the file path
file_path = r'F:\MLpractrice\weather1\data\modified_rainfall.csv'

# Read the CSV file into a DataFrame
rainfall_data = pd.read_csv(file_path)

# Replace any occurrence of '*' with NaN and then NaN with 0
rainfall_data.replace(r'\*+', np.nan, regex=True, inplace=True)
rainfall_data.iloc[:, 3:] = rainfall_data.iloc[:, 3:].apply(pd.to_numeric, errors='coerce')
rainfall_data.fillna(0, inplace=True)

# Example 1: Plotting total rainfall for each month for a specific station (e.g., 'Bogra')
station = 'Bogra'
station_data = rainfall_data[rainfall_data['Station'] == station]
monthly_rainfall = station_data.groupby('Month').sum().iloc[:, 3:].sum(axis=0)

# Plotting the total rainfall by month for 'Bogra'
plt.figure(figsize=(10, 6))
monthly_rainfall.plot(kind='bar', color='skyblue')
plt.title(f"Monthly Rainfall for {station}")
plt.xlabel("Month")
plt.ylabel("Total Rainfall (mm)")
plt.xticks(rotation=0)
plt.show()

# Example 2: Heatmap of the entire dataset showing rainfall for all stations, months, and days
# We will aggregate data to show the rainfall per station per month.
rainfall_by_station_month = rainfall_data.groupby(['Station', 'Month']).sum().iloc[:, 3:]

# Pivot the data to get stations as rows and months as columns
rainfall_pivot = rainfall_by_station_month.unstack(level='Month').fillna(0)

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(rainfall_pivot, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Rainfall (mm)'})
plt.title("Rainfall Heatmap by Station and Month")
plt.xlabel("Month")
plt.ylabel("Station")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Specify the file path
file_path = r'F:\MLpractrice\weather1\data\modified_rainfall.csv'

# Read the CSV file into a DataFrame
rainfall_data = pd.read_csv(file_path)

# Replace any occurrence of '*' with NaN and then NaN with 0
rainfall_data.replace(r'\*+', np.nan, regex=True, inplace=True)
rainfall_data.iloc[:, 3:] = rainfall_data.iloc[:, 3:].apply(pd.to_numeric, errors='coerce')
rainfall_data.fillna(0, inplace=True)

# Filter data for 'Rajshahi' station
station = 'Rajshahi'
rajshahi_data = rainfall_data[rainfall_data['Station'] == station]

# Calculate total rainfall by month (sum across days for each month)
monthly_rainfall = rajshahi_data.groupby('Month').sum().iloc[:, 3:].sum(axis=0)

# Plotting the pie chart for Rajshahi's total rainfall by month
plt.figure(figsize=(8, 8))
monthly_rainfall.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, legend=False)
plt.title(f"Monthly Rainfall Distribution for {station}")
plt.ylabel('')  # Remove the 'ylabel' to make it cleaner
plt.show()
```
