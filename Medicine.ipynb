```python
import pandas as pd

# Specify the path to the CSV file
file_path = r'F:\MLpractrice\medicine\data\DIRECTORATE_GENERAL_OF_DRUG_ADMINISTRATION\DIRECTORATE_GENERAL_OF_DRUG_ADMINISTRATION.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to confirm it's loaded correctly
df.head()
```

```python
import matplotlib.pyplot as plt

# Get the value counts for the 'Dosages Description' column
dosage_description_counts = df['Dosages Description'].value_counts()

# Plot the column chart (bar chart)
plt.figure(figsize=(10, 6))  # Adjust the size for better visualization
dosage_description_counts.plot(kind='bar', color='skyblue')

# Title and labels
plt.title('Distribution of Dosages Description', pad=20)
plt.xlabel('Dosages Description')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.tight_layout()  # To prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the value counts for the 'Dosages Description' column
dosage_description_counts = df['Dosages Description'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))

# Create the pie chart
plt.pie(dosage_description_counts, 
        labels=dosage_description_counts.index, 
        autopct='%1.1f%%',  # Show percentage on the chart
        startangle=90, 
        colors=plt.cm.Paired.colors,  # Choose a color palette
        pctdistance=0.85,  # Moves the percentage closer to the edge
        textprops={'fontsize': 10})  # Set font size for the labels

# Title for the pie chart with padding
plt.title('Distribution of Dosages Description', pad=20)

# Equal aspect ratio ensures that pie chart is circular
plt.axis('equal')

# Display the pie chart
plt.show()
```
